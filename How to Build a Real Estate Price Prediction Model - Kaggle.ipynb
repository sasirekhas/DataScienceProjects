```python
# there are  lots of columns with skewed data whitch make it hard to catch them so lets try check it with if statement and see what will happen
```

```python
# select columns with skew()>1 or <1

sk=[]
for i in df.drop(columns=[ 'YearBuilt','YearRemodAdd']).select_dtypes('number').columns:
    if ((df[i].skew()>1) or (df[i].skew()<1)):
        sk.append(i)
sk
```

```python
# update values of skew data from x to log(x)

sk_=pd.DataFrame(np.select([DF[sk]==0, DF[sk] > 0, DF[sk] < 0], [0, np.log(DF[sk]), np.log(DF[sk])]),columns=sk).set_index(DF.index)
df_skew=DF.drop(columns=sk).set_index(DF.index)

df_skew=pd.concat([df_skew,sk_],axis=1)
X_train_skew=df_skew.drop(columns='SalePrice')
```

```python
# select columns with skew()>1 or <1

sk_t=[]
for i in test.drop(columns=[ 'YearBuilt','YearRemodAdd']).select_dtypes('number').columns:
    if ((test[i].skew()>1) or (test[i].skew()<1)):
        sk_t.append(i)
sk_t=pd.DataFrame(np.select([test[sk_t]==0, test[sk_t] > 0, test[sk_t] < 0], [0, np.log(test[sk_t]), np.log(test[sk_t])]),columns=sk_t).set_index(test.index)
df_skew_t=test.drop(columns=sk_t).set_index(test.index)

df_skew_t=pd.concat([df_skew_t,sk_t],axis=1)
X_test_skew=df_skew_t.reindex(columns=X_train_skew.columns)
X_test_skew.shape
```

Impute missing values
imputing missing values for categorical features is diff from numerical features in strategy

so lets split the data for cat and num features and impute each one in separate imputer

```python
# start with Categorical Features
imp_cat = SimpleImputer(strategy="most_frequent")

X_train_cat=X_train.select_dtypes('object')
X_test_cat=X_test.select_dtypes('object')
X_train_cat = imp_cat.fit_transform(X_train_cat)
X_test_cat = imp_cat.transform(X_test_cat)
```

```python
# Numerical Features
imp_num = SimpleImputer()

X_train_num=X_train.select_dtypes('number')
X_test_num=X_test.select_dtypes('number')
X_train_num = imp_num.fit_transform(X_train_num)
X_test_num = imp_num.transform(X_test_num)
# Conatinate twice to get them pack as last version 
X_train_cat=pd.DataFrame(X_train_cat,columns=X_train.select_dtypes('object').columns)
X_train_num=pd.DataFrame(X_train_num,columns=X_train.select_dtypes('number').columns)
X_train=pd.concat([X_train_cat,X_train_num],axis=1).set_index(DF.index)
```

```python
X_test_cat=pd.DataFrame(X_test_cat,columns=X_test.select_dtypes('object').columns)
X_test_num=pd.DataFrame(X_test_num,columns=X_test.select_dtypes('number').columns)
X_test=pd.concat([X_test_cat,X_test_num],axis=1)
X_test=X_test.reindex(columns=X_train.columns)
```

```python
#Fitting a model and Hyperparameter tunin
```

```python
model_names=["ridge" ,"lasso", "regressor"]
models=[Ridge(),Lasso(),LinearRegression()] # to put model in loop to get more variety of results

# params of each model 

param_ridge={
        'ridge__alpha':np.arange(0.0001,1,10),
        'ridge__solver':["auto", "svd", "cholesky", "lsqr", "sparse_cg", "sag", "saga"]

}

param_lasso={
        'lasso__alpha':np.arange(0.01,50.0,20)
}
param_regressor={
                 }


# dicts to get pack best model , pest params and lowest Error

best_params={}
result={}
RMSE={}
score={}
for name,model in  zip(model_names,models):
    steps = [
            ("onehot", OneHotEncoder()),  
             ('scaler',StandardScaler()),
            (name, model)]
    # Pipe line for using OneHotEncoder and StandardScaler 
    pipeline = Pipeline(steps) 

    if name == 'regressor' :
        print("regressor")
        cv_regressor=GridSearchCV(pipeline,param_grid=param_regressor)
        cv_regressor.fit(X_train,y_train)
        y_pred_train=cv_regressor.predict(X_train)
        RMSE[name]=mean_squared_error(y_train,y_pred_train,squared=False)
        best_params[name]=cv_regressor.best_params_
        result[name]=cv_regressor.predict(X_test)
        score[name]=cv_regressor.score(X_train,y_train)
    elif name == 'ridge':
        cv_ridge=GridSearchCV(pipeline,param_grid=param_ridge)
        cv_ridge.fit(X_train,y_train)
        y_pred_train=cv_ridge.predict(X_train)
        
        RMSE[name]=mean_squared_error(y_train,y_pred_train,squared=False)
        best_params[name]=cv_ridge.best_params_
        result[name]=cv_ridge.predict(X_test)
        score[name]=cv_ridge.score(X_train,y_train)
    else:
        cv_lasso=GridSearchCV(pipeline,param_grid=param_lasso)
        cv_lasso.fit(X_train,y_train)
        y_pred_train=cv_lasso.predict(X_train)

        RMSE[name]=mean_squared_error(y_train,y_pred_train,squared=False)
        best_params[name]=cv_lasso.best_params_
        result[name]=cv_lasso.predict(X_test)
        score[name]=cv_lasso.score(X_train,y_train)
print('Done')
```
