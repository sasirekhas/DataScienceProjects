```python
import seaborn as sns
```

```python
import matplotlib.pyplot as plt   #defines the plt alias for the pyplot module
```

```python
def cheks_nan(data):
    a = data.isnull().sum()
    a = a[a>0].sort_values()
    index_ = a.index
    nan_data = pd.DataFrame(data=[index_,a], index=['Name_Nan', 'Count_Nan']).transpose()
    nan_data['Count_Nan'] = nan_data['Count_Nan'].apply(lambda x: int(x))
    sns.barplot(data=nan_data, x='Name_Nan', y='Count_Nan');
    plt.ylim(0,1500);
    plt.xticks(rotation=90);
```

```python
cheks_nan(train_data)
```
